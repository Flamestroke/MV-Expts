```python
# HOG (Histogram of Oriented Gradients) descriptor

import cv2
import numpy as np
from matplotlib import pyplot as plt

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

img = cv2.imread('peeps.png')

boxes, weights = hog.detectMultiScale(img, 
                                      winStride=(4, 4),
                                      padding=(3, 4), 
                                      scale=1.57)

for (x, y, w, h) in boxes:
    cv2.rectangle(img, 
                  (x, y), 
                  (x+w, y+h), 
                  (0, 255, 255), 
                  7)
    
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# Person detection using HOGDescriptor

img = cv2.imread('walk.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def is_inside(o, i):
    ox, oy, ow, oh = o
    ix, iy, iw, ih = i
    return ox < ix and oy < iy and ox + ow > ix + iw and oy + oh > iy + ih

def draw_person(image, person):
    x, y, w, h = person
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 255, 0), 8)
boxes, weights = hog.detectMultiScale(img, 
                                      winStride=(4, 4),
                                      padding=(10, 4), 
                                      scale=1.05)
found_filtered = []

for ri, r in enumerate(boxes):
     for qi, q in enumerate(boxes[:ri]):
        if is_inside(r, q):
            break
     else:
        found_filtered.append(r)
for person in found_filtered:
    draw_person(img, person)
    
plt.imshow(img)
plt.axis('off')
```
